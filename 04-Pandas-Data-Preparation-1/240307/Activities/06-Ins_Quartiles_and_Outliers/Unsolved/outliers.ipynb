```python
# Dependencies
import pandas as pd
```

```python
# Read in the LAX temperatures dataset and create a box plot
temp_df = pd.read_csv("../Resources/lax_temperature.csv")
temperature = temp_df['HourlyDryBulbTemperature']

temp_df
```

```python
# Use pandas to find potential outliers by calculating the interquartile range (IQR)


Q1 = temperature.quantile(0.25)


Q2 = temperature.quantile(0.5)


Q3 = temperature.quantile(0.75)


IQR = Q3 - Q1

print(f"The lower quartile of temperatures is: {Q1}")
print(f"The upper quartile of temperatures is: {Q3}")
print(f"The interquartile range of temperatures is: {IQR}")
print(f"The the median of temperatures is: {Q2} ")
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# Create a DataFrame of rows that could be outliers
outliers = temp_df.loc[(temp_df['HourlyDryBulbTemperature'] < lower_bound) | 
                       (temp_df['HourlyDryBulbTemperature'] > upper_bound)]

outliers


non_outliers = temp_df.loc[~(temp_df['HourlyDryBulbTemperature'] < lower_bound) | 
                            (temp_df['HourlyDryBulbTemperature'] > upper_bound)]
non_outliers
```
