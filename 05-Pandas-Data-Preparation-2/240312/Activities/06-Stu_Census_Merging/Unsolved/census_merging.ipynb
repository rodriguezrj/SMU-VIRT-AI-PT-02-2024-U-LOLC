```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store filepaths into variable
state_avg_csv = Path("Resources/state_avg.csv")
state_totals_csv = Path("Resources/state_totals.csv")
```

```python
# Read in files.
state_avg = pd.read_csv(state_avg_csv)
state_totals = pd.read_csv(state_totals_csv)
```

```python
# Display the state averages.
print(state_avg.shape)
state_avg.head()
```

```python
# Display the state totals.
print(state_totals.shape)
state_totals.head()
```

```python
# Merge the two DataFrames together based on the Year and State they share
census_df = pd.merge(state_avg, state_totals, on = ["Year", "State"])

census_df
```

```python
census_df['Year'].value_counts()
```

```python
# Create a DataFrame that filters the data on only 2019
# census_df.loc[census_df['Year']==2019]
census_2019 = census_df.loc[census_df['Year']==2019, :]
census_2019.head()
```

```python
# Add a new column that calculates the Poverty Rate
census_2019['Poverty Rate (%)'] = (census_2019['Total Population in Poverty'] / 
                                   census_2019['Total Population']) * 100
census_2019.head()

# census_2019.loc[:, 'Poverty Rate (%)'] = (census_2019['Total Population in Poverty'] / 
#                                          census_2019['Total Population']) * 100
# census_2019.head()
```

```python
# Sort the data by Poverty Rate and Average Per Capita Income by County, Highest to Lowest
census_2019  = census_2019.sort_values(by=['Poverty Rate (%)', 'Average Per Capita Income by County'], ascending=False)
census_2019
# Reset Index
sorted_census = census_2019.reset_index(drop=True)
sorted_census
```

```python
# Print out the data for the state or territory with the highest poverty rate
sorted_census.iloc[0, :]
```

```python
# Print out the data for the state or territory with the lowest poverty rate with one line of code
sorted_census.iloc[-1, :]
```
