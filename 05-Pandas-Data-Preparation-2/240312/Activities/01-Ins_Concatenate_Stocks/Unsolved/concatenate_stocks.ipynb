#### Import Libraries and Dependencies 

```python
import pandas as pd
from pathlib import Path
```

#### Read in Files 

```python
# Import data

apple_path = Path('Resources/AAPL.csv')
google_path = Path('Resources/GOOG.csv')
meta_path = Path('Resources/META.csv')

# Read in data and index by date
apple_data = pd.read_csv(apple_path, index_col='Date')
goog_data = pd.read_csv(google_path, index_col='Date')
meta_data = pd.read_csv(meta_path, index_col='Date')
```

#### Output of sample data

```python
# Show sample of Apple data
print(apple_data.shape)
apple_data
```

```python
# Show sample of Google data
print(goog_data.shape)
goog_data
```

```python
# Show sample of Meta data
print(meta_data.shape)
meta_data
```

#### Concatenate data by rows using concat function and inner join

```python
# Join by rows axis
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=0, join="inner") # axis=rows

joined_data_rows.head(30)
```

#### Concatenate data by columns using concat function and inner join

```python
joined_data_rows.index
```

```python
# Join by columns axis
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=1, join="inner") # axis="columns"

joined_data_rows.head(30)
```

#### Concatenate data by rows using concat function and inner join and provide the stock as keys.

```python
# Join by rows and add the stock ticker as the key.
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=0, join="inner",
                             keys=['Apple', "Google", "Metaverse"]) # axis=rows

joined_data_rows.head(30)
```

```python
apple_data.index = apple_data.index.rename("date")
```

```python
apple_data = apple_data.set_index(apple_data["Open"])
```

```python
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=0, join="inner",
                            keys=['Apple', "Google", "Metaverse"]) # axis=rows

joined_data_rows.head(30)
```

```python
joined_data_rows
```

#### Combine the DataFrames horizontally using the `axis="columns"`.

```python
# Join by columns and add the stock ticker as the key.
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=1, join="inner",
                            keys=['Apple', "Google", "Metaverse"]) # axis=rows

joined_data_rows.head(30)
```

```python
# Join by columns and add the stock ticker as the key.
joined_data_rows = pd.concat([apple_data, goog_data, meta_data], axis=1, join="outer",
                            keys=['Apple', "Google", "Metaverse"]) # axis=rows

joined_data_rows.head(30)
```

## Example

```python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
df2 = pd.DataFrame({'B': [7, 8, 9], 'C': [10, 11, 12]}, index=['b', 'c', 'd'])
```

```python
df1
```

```python
df2
```

### with columns

```python
concatenated_df = pd.concat([df1, df2], axis="columns", join="inner")

concatenated_df
```

```python
concatenated_df = pd.concat([df1, df2], axis="columns", join="outer")

concatenated_df
```

### with rows

```python
concatenated_df = pd.concat([df1, df2], axis="rows", join="inner")
concatenated_df
```

```python
concatenated_df = pd.concat([df1, df2], axis="rows", join="outer")
concatenated_df
```
