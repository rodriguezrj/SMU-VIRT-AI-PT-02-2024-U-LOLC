```python
# Import necessary libraries and dependencies
import numpy as np
import pandas as pd
```

```python
# Import data
sp500_path = '../Resources/sp500_stock_volume.csv'

# Read the S&P 500 volume into a DataFrame. (Make sure to declare the datetime index)
sp500 = pd.read_csv(sp500_path, index_col="Date", parse_dates=True)

# Display sample data
sp500.head()
```

```python
# Slice the dataframe so that it just includes the volume data.
vol = sp500['volume']
```

```python
# Declare the group level to be the day of the week (e.g., Mon, Tues,...)


# Plot average daily volume according to day of the week
vol.groupby([vol.index.dayofweek]).mean().plot(style='g--')
```

```python
# Declare the group level to be hour of the day


# Plot average hourly volume
vol.groupby([vol.index.hour]).mean().plot(style='g--')
```

```python
# Group the data by the calendar week in the year (week of year). 
vol.groupby(by=[vol.index.year, vol.index.strftime("%U")]).mean().plot(
title="SP500 Mean Volume per Week", figsize=(10, 5))
```

```python
vol.groupby(by=[vol.index.year, vol.index.isocalendar().week]).mean().plot(
title="SP500 Mean Volume per Week", figsize=(10, 5))
```
