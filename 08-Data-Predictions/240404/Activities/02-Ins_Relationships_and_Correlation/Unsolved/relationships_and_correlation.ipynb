```python
# Initial imports
import numpy as np
import pandas as pd
import hvplot.pandas
```

```python
# Set the file path
file_path = "../Resources/national-home-sales.csv"
```

```python
# Load time series data into Pandas
df_home_sales = pd.read_csv(
    file_path,
    index_col="period_end_date",
    parse_dates=True
)
```

```python
# Display sample data
df_home_sales.head(10)
```

```python
# Sort the DataFrame index in ascending order
df_home_sales = df_home_sales.sort_index()
# Display sample data
df_home_sales.head(10)
```

```python
df_home_sales.plot()
```

```python
# Plot the inventory and homes_sold series
df_home_sales[['homes_sold', 'inventory']].plot()
```

```python
scatter_plot = df_home_sales.hvplot.scatter(x='inventory', y='homes_sold', title='Inventory vs. Homes Sold')
scatter_plot
```
