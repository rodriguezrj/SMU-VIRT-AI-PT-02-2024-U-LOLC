```python
# Dependencies
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
# Read in the wine recognition data set from sklearn and load into Pandas
data = dta.load_wine()
wine_data = pd.DataFrame(data.data,columns=data.feature_names)
print(data.DESCR)
```

```python
# Plot flavanoids versus malic_acid on a scatterplot
flav  = wine_data['flavanoids']
malic = wine_data['malic_acid']
plt.scatter(malic, flav)
plt.xlabel("Amount of Malic Acid")
plt.ylabel("Amount of Flavanoids")
plt.show()
```

```python
# Calculate the correlation coefficient between malic_acid and flavanoids
corr = round(st.pearsonr(flav, malic)[0], 2)
print(f"The correlation coefficient between malic acid and flavanoids is {corr}")
```

```python
# Plot colour_intensity versus alcohol on a scatterplot
col = wine_data['color_intensity']
alc = wine_data['alcohol']
plt.scatter(alc, col)
plt.xlabel('Amount Alc')
plt.ylabel('Intensity of Color')
plt.show()
```

```python
# Calculate the correlation coefficient between alcohol and color_intensity
corr = round(st.pearsonr(col, alc)[0], 2)
print(f"The correlation coefficient between alcohol and color intensity is {corr}")
```
