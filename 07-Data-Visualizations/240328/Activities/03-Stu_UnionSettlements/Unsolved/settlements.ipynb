```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
df = pd.read_csv('./Resources/union_settlements_1995.csv')

df.head()
```

```python
# List columns
df.columns
```

```python
# Get total settlements by union
data = df['UNION'].value_counts()
data
```

```python
# Configure plot, figsize, title, and axis labels
data_viz = data.plot(kind='bar', figsize = (8,8), 
                     title="Union Settlements", 
                     x="Union", y="Num settlements")

# Configure x-tick rotation
data_viz.set_xticklabels(data.index, rotation=45, 
                         rotation_mode="anchor", ha='right', 
                         wrap=True) 

# Show plot
plt.show()
```
